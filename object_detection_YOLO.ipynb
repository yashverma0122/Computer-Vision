```python
import cv2
import numpy as np

# Load YOLOv4 configuration and weights files
net = cv2.dnn.readNet("/home/lookman/darknet/cfg/yolov4.cfg", "/home/lookman/darknet/yolov4.weights")

# Specify classes of objects to detect
classes = []
with open("/home/lookman/darknet/data/coco.names", "r") as f:
    classes = [line.strip() for line in f.readlines()]

# Set minimum confidence threshold for object detection
conf_threshold = 0.5

# Set non-maximum suppression threshold
nms_threshold = 0.4

new_width = 720
new_height = 480

# Load video
cap = cv2.VideoCapture("/home/lookman/Desktop/yash/new/image_examples/gun1.mp4")
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

# Calculate the scaling factors for the new dimensions
width_scale = new_width / width
height_scale = new_height / height

# Create a resizing function
def resize(frame):
    return cv2.resize(frame, (new_width, new_height))

# # Define output video writer object
result = cv2.VideoWriter('final.avi',cv2.VideoWriter_fourcc(*'MJPG'),30.0,(new_width, new_height))


# Loop through frames in video
while True:
    # Read frame
    ret, frame = cap.read()
    if not ret:
        break
        
     # Resize frame
    frame = resize(frame)

    # Create blob from frame
    blob = cv2.dnn.blobFromImage(frame, 1/255, (416, 416), (0, 0, 0), swapRB=True, crop=False)

    # Set input to network
    net.setInput(blob)

    # Get output from network
    output_layers_names = net.getUnconnectedOutLayersNames()
    layer_outputs = net.forward(output_layers_names)

    # Process detections
    boxes = []
    confidences = []
    class_ids = []
    for output in layer_outputs:
        for detection in output:
            scores = detection[5:]
            class_id = np.argmax(scores)
            confidence = scores[class_id]
            if confidence > conf_threshold:
                center_x = int(detection[0] * frame.shape[1])
                center_y = int(detection[1] * frame.shape[0])
                width = int(detection[2] * frame.shape[1])
                height = int(detection[3] * frame.shape[0])
                left = int(center_x - width / 2)
                top = int(center_y - height / 2)
                boxes.append([left, top, width, height])
                confidences.append(float(confidence))
                class_ids.append(class_id)

    # Apply non-maximum suppression
    indices = cv2.dnn.NMSBoxes(boxes, confidences, conf_threshold, nms_threshold)

    # Draw boxes and labels around detected objects
    if len(indices) > 0:
        for i in indices.flatten():
            x, y = boxes[i][0], boxes[i][1]
            w, h = boxes[i][2], boxes[i][3]
            color = [int(c) for c in np.random.uniform(0, 255, size=(3,))]
            cv2.rectangle(frame, (x, y), (x + w, y + h), color, 2)
            text = f"{classes[class_ids[i]]}: {confidences[i]:.2f}"
            cv2.putText(frame, text, (x, y - 5), cv2.FONT_HERSHEY_SIMPLEX, 0.5, color, 2)
    
    
    # Write output frame to video
    if ret == True:
        result.write(frame)
#         if cv2.waitKey(1) == 13:
#             break
    else:
        print('some error occured')
        break
            


#     Show frame with detected objects
    cv2.imshow("Object detection using YOLOv4", frame)
    if cv2.waitKey(1) == 13:
        break



# Release video capture and destroy all windows
# cap.release()
result.release()
cv2.destroyAllWindows()
print('video saved succesfully')
```
